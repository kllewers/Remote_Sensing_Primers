Remote Sensing Primer: Synthetic Aperture Radar (SAR) Satellites

## 1. Why Should Biodiversity Researchers Care About SAR?

Synthetic Aperture Radar (SAR) offers a fundamentally different lens for observing biodiversity-related phenomena compared to traditional optical remote sensing. Because SAR is an **active sensor** — meaning it emits its own microwave signal and records the reflection — it can operate **independently of daylight** and **penetrate cloud cover**. This makes it especially valuable in the tropics, polar regions, or any location with frequent cloud obstruction.

Beyond mere visibility, SAR captures **structural properties of vegetation and terrain**. It is sensitive to things like surface roughness, moisture content, and vegetation density, all of which are important ecological indicators. SAR time-series can also be used to detect **seasonal dynamics**, **inundation patterns**, or **land surface deformation**, enriching our ability to interpret species occurrences and distributions in space and time.

In a biodiversity informatics context, SAR can:

* Provide high-resolution, consistent environmental context across time for areas where species are observed.
* Enable monitoring of **habitat quality**, **deforestation**, **wetland inundation**, and **agricultural expansion**.
* Support **species distribution modeling** with continuous and cloud-independent environmental covariates.

## 2. How SAR Complements GBIF & Biodiversity Data

Most GBIF datasets are occurrence-based: they tell us where and when a species was recorded. But they rarely tell us **why** that species is there — what ecological factors supported its presence at that location.

SAR helps fill that gap by providing dynamic environmental data:

* **Habitat Monitoring**: Monitor floodplains, forest canopy recovery, and human-driven habitat change with SAR backscatter over time.
* **Disturbance Detection**: Use SAR to detect landslides, fires, or clear-cutting events that affect species presence or abundance.
* **Moisture Dynamics**: SAR can sense surface wetness and moisture, helping to interpret records for amphibians, insects, or plants in ephemeral habitats.

A potential use case: combining GBIF’s occurrence data for wetland bird species with Sentinel-1 SAR imagery to detect surface water extent and seasonal flooding that coincide with those sightings.

## 3. A Quick SAR Primer for the Uninitiated

SAR is different from the passive optical imagery many biodiversity researchers are used to (like Landsat or Sentinel-2). It operates in the **microwave region** of the electromagnetic spectrum, meaning it can observe Earth’s surface **regardless of lighting or cloud cover**.

### How It Works

* A SAR sensor emits a microwave pulse toward the Earth’s surface.
* That pulse interacts with the ground (vegetation, soil, water, built structures), and some of the energy is scattered back to the sensor.
* The strength and timing of the returned signal reveal properties of the surface.

### SAR Bands & Biodiversity Use

* **L-band (1–2 GHz)**: Penetrates deeper into vegetation, useful for forest biomass, root-zone soil moisture.
* **C-band (4–8 GHz)**: Good balance between vegetation sensitivity and resolution. Sentinel-1 operates in this band.
* **X-band (8–12 GHz)**: Higher resolution, sensitive to surface features — used in urban and coastal mapping.

### Polarization

SAR signals can be transmitted and received in different polarizations:

* **HH**: Horizontal transmit, horizontal receive
* **VV**: Vertical transmit, vertical receive
* **HV / VH**: Cross-polarized

These options help distinguish vegetation types, surface roughness, and scattering mechanisms.

### Key SAR Satellites (Open Access Focus)

* **Sentinel-1 (ESA)**: C-band SAR with 6–12 day revisit, freely available, global coverage.
* **ALOS-2 (JAXA)**: L-band SAR, useful for biomass and forest structure studies.
* **SAOCOM (Argentina)**: L-band SAR, growing open data support.
* **NISAR (NASA/ISRO)**: Launching soon with dual L- and S-band coverage and full open access.

These missions provide rich datasets that biodiversity researchers can integrate into their workflows with relatively low technical barriers.
